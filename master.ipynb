```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('csv_files/Prime_Ages.csv')
```

```python
df_zoomed = df[(df['Average_Prime_Age'] >= 24) & (df['Average_Prime_Age'] <= 27)]

plt.figure(figsize=(10, 6))
plt.bar(df_zoomed['Position'], df_zoomed['Average_Prime_Age'])
plt.xlabel('Position')
plt.ylabel('Age')
plt.title('Average Prime Age by Position in NFL')
plt.xticks(rotation=45)
plt.ylim(24, 27)
plt.tight_layout()

plt.show()
```

```python
df_zoomed = df_zoomed.sort_values(by='Average_Prime_Age')

plt.figure(figsize=(10, 6))
plt.bar(df_zoomed['Position'], df_zoomed['Average_Prime_Age'])
plt.xlabel('Position')
plt.ylabel('Age')
plt.title('Average Prime Age by Position in NFL')
plt.xticks(rotation=45)
plt.ylim(24, 27)
plt.tight_layout()

plt.show()
```
